## Data analysis

Write a function that prints the mean and another that prints the standard deviation of each attribute.

```python
import numpy as np

# Load and clean iris.data manually (this part works!)
with open("iris.data", "r") as f:
    lines = [line.strip() for line in f if line.strip()]

rows = [line.split(",") for line in lines]
X = np.array([row[:4] for row in rows], dtype=float)
y = np.array([row[4] for row in rows], dtype=str)

# Attribute name map
attribute_names = {
    0: "sepal length",
    1: "sepal width",
    2: "petal length",
    3: "petal width"
}

# Function: print means
def print_attribute_means(X, attribute_names):
    print("Attribute Means:")
    for i in range(X.shape[1]):
        mean_val = np.mean(X[:, i])
        print(f"{attribute_names[i]}: {mean_val:.2f}")

# Function: print standard deviations
def print_attribute_stddevs(X, attribute_names):
    print("Attribute Standard Deviations:")
    for i in range(X.shape[1]):
        std_val = np.std(X[:, i])
        print(f"{attribute_names[i]}: {std_val:.2f}")

# Run them!
print_attribute_means(X, attribute_names)
print()
print_attribute_stddevs(X, attribute_names)
```

Use the functions you defined above to print the mean of the attributes for each class.

```python
print(print_attribute_means(X, attribute_names))
```

Now, let's plot this data. Use `matplotlib` to create scatter plots for petal and sepal. Use different colors for each class.

```python
import matplotlib.pyplot as plt
# Define class names and colors
classes = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['red', 'green', 'blue']
plt.figure(figsize=(8, 6))
for class_name, color in zip(classes, colors):
    mask = y == class_name
    plt.scatter(X[mask, 0], X[mask, 1], label=class_name, color=color)
    
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Sepal Width")
plt.legend()
plt.grid(True)
plt.show()
plt.figure(figsize=(8, 6))
for class_name, color in zip(classes, colors):
    mask = y == class_name
    plt.scatter(X[mask, 2], X[mask, 3], label=class_name, color=color)
    
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal Length vs Petal Width")
plt.legend()
plt.grid(True)
plt.show()
```
